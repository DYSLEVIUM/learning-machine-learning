# K-Mean Clustering

K-means is a type of unsupervised learning. K-means algorithm identifies k number of centroids and then allocates every data point to the nearest cluster while keeping the centroids as small as possible. The 'means' in the K-means refers to the averaging of the data, i.e., finding the centroid.

# Types of Clustering

1. Hard Clustering: Each data point either belongs to a class or not. (mutually exclusive)

2. Soft clustering: One data point can belong to more than one cluster with some probability of likelihood.

# Types of Clustering Algoritms

  1. Connectivity-based clustering: The main idea behind this cludtering is that data points that are closed in the data space are more related(similar) than to data points farther away. They are also not very robust towards outliers, which might show up as additional clusters or even cause other clusters to merge.

<center><img src="https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Ftse3.mm.bing.net%2Fth%3Fid%3DOIP.E3tWfHTsXPEqFp1H7tpbYwAAAA%26pid%3DApi&f=1"/></center>

  2. Centroid-based clustering: In this type of clustering, clusters are represented by a central vector or a centroid. This centroid might not neccessarily be a member of the dataset.

<center><img src="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQTOOyZJ7nCxtxXni-OzfM-xCMDt9sNE7pm-OADahaGp-Ymxk_qaTelf4RhZUede9mA8N8&usqp=CAU"/></center>

  3. Distribution-based clustering: These models have a strong theoretical foundation, however they ofter suffer from overfitting. Gaussian mixture models, using the expectation-maximization algorithm is a famous distribution based clustering method.

  4. Density-based clustering: These methods search the data space for areas of varied density of data points.

# Working with K-means Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
data=pd.read_csv('data.csv', index_col=0)
data.head()
```

```python
data['cluster'].value_counts()
```

```python
plt.scatter(data['x'], data['y'], c=data['cluster'], cmap="viridis")
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X=data[['x', 'y']]
y=data['cluster']
```

```python
X
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
data[['x', 'y']] = X
```

```python
kmeans=KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
plt.scatter(data['x'], data['y'], c=kmeans.labels_, cmap="viridis")
```

# How to choose the right vakue of clusters

Assuming we have inputs x_1, x_2, x_3, , x_nx

    Step 1  Pick K random points as cluster centers called centroids.
    Step 2  Assign each x_i to nearest cluster by calculating its distance to each centroid.
    Step 3  Find new cluster center by taking the average of the assigned points.
    Step 4  Repeat Step 2 and 3 until none of the cluster assignments change.

<center><img src="https://i.imgur.com/k4XcapI.gif"/></center>

# Elbow Method
We often know the value of K. In that case we use the value of K. Else we use the Elbow Method.
<center><img src="https://kgptalkie.com/wp-content/uploads/2020/08/image-138.png"/></center>

# Error Sum of Squares (SSE)

It is the sum of the squared differences between each observation and its groups mean. We can use it as a measure of variation within a cluster. All cases within a cluster are identical then SSE would be equal to 0. We run the algorithm for different values of K(say K = 10 to 1) and plot the K values against SSE(Sum of Squared Errors). And select the value of K for the elbow point as shown in the figure.

<center><img src="https://kgptalkie.com/wp-content/ql-cache/quicklatex.com-63845549b8c98061fbd2be8b1d3f121b_l3.svg"/></center>


```python
SSE = []
index=range(1,10)
for i in index:
  kmeans=KMeans(n_clusters=i, random_state=42)
  kmeans.fit(X)
  SSE.append(kmeans.inertia_)
  print(i, kmeans.inertia_)
```

```python
plt.plot(index, SSE)
plt.xlabel('K')
plt.ylabel('SEE')
plt.title('SSE with respect to K')
plt.show()
```

```python
k=3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)
center = kmeans.cluster_centers_
plt.xlabel('X-values')
plt.ylabel('Y-values')
plt.title('Formation of clusters with 3-centroids')

plt.scatter(data['x'], data['y'], c=kmeans.labels_, cmap = 'viridis')
for i, point in enumerate(center):
  plt.plot(center[i][0], center[i][1], '*r--', linewidth=2, markersize=18)
```
